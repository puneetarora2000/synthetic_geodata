```python
import pandas as pd
import numpy as np

# Define parameters
n = 1000  # number of users
dn = 30  # number of days
activities = [1, 2, 3, 4, 5, 6]  # 1=walk, 2=running, 3=driving, 4=sitting, 5=shopping, 6=sleeping
locations = {'park': 1, 'mall': 2, 'office': 3, 'home': 4}
lat_min, lat_max = 35.0, 40.0
lon_min, lon_max = -120.0, -80.0

# Generate data
users = np.repeat(np.arange(1, n+1), dn*6)
days = np.tile(np.repeat(np.arange(1, dn+1), 6), n)
activities = np.tile(activities, n*dn)
locations = np.random.choice(list(locations.values()), n*dn*6)
lat = np.random.uniform(lat_min, lat_max, n*dn*6)
lon = np.random.uniform(lon_min, lon_max, n*dn*6)

# Combine into dataframe
df = pd.DataFrame({'user_id': users,
                   'day': days,
                   'activity': activities,
                   'location': locations,
                   'lat': lat,
                   'lon': lon})

# Convert location codes back to names
df['location'] = df['location'].map({v: k for k, v in locations.items()})

# Print first few rows of dataset
print(df.head())
```

```python
!pip install geopy folium

from geopy.distance import geodesic
import random
import folium

def generate_synthetic_geodata(city_coords, radius_km, num_points):
    activities = ['cycling', 'driving', 'standing', 'sleeping', 'walking']
    synthetic_geodata = []
    for _ in range(num_points):
        angle = random.uniform(0, 360)
        distance = random.uniform(0, radius_km)
        activity = random.choice(activities)
        new_coords = geodesic(kilometers=distance).destination(point=city_coords, bearing=angle)
        synthetic_geodata.append((new_coords.latitude, new_coords.longitude, activity))
    return synthetic_geodata

def create_and_display_map(center_coords, geodata):
    m = folium.Map(location=center_coords, zoom_start=12)
    for coord in geodata:
        latitude, longitude, activity = coord
        folium.Marker(
            location=(latitude, longitude),
            popup=f"Activity: {activity}"
        ).add_to(m)
    return m

# Customize parameters
city_coords = (37.7749, -122.4194)  # San Francisco
radius_km = 7
num_points = 100

# Generate synthetic geodata
synthetic_geodata = generate_synthetic_geodata(city_coords, radius_km, num_points)

# Create and display the map
m = create_and_display_map(city_coords, synthetic_geodata)
m
```
